```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("IPL_2025_Player_Statistics.csv")
```

```python
df.tail()# shows the last five data
```

```python
df.info()
```

```python
df.describe()
```

```python
run_array = np.array(df['Runs'])
wickets_array = np.array(df['Wickets'])
print('Mean of Runs',np.mean(run_array))
print("standard deviation of wickets",np.std(wickets_array))
```

```python
#series is a 1d array capable of holdiing a lot of elements
#we need to find out the players who score above 1000 runs
high_scores=df[df['Runs']>1000]
high_scores[['Player','Team','Runs']]
```

```python
print(high_scores.size)
```

```python
#arange or group the teams with highest to loest wicket taking
team_stats = df.groupby('Team')[['Runs','Wickets']].sum()
team_stats.sort_values(by='Wickets',ascending=False)
```

```python
team_stats['Runs'].plot(kind='bar',color='skyblue')
plt.title('Runs Scored By each team and Wickets taken also.')
plt.xlabel('Teams')
plt.ylabel('Runs')
plt.grid(True)
plt.show()
```

```python
role_counts = df['Player Role'].value_counts()
plt.pie(role_counts,labels = role_counts.index,autopct='%1.1f%%',startangle=140)
plt.title('player Role Distribution')
plt.axis('Equal')
plt.show()
```

```python
#histogram
plt.hist(df['Strike Rate'], color='orange',edgecolor = 'black')
plt.title('Histrogram of batting Strike Rate')
plt.xlabel('Strike Rate')
plt.ylabel('Number of Wicket')
plt.grid(True)
plt.show()
```

```python
#scatter plot
plt.scatter(df['Runs'],df['Wickets'],color='green')
plt.title("Runs vs Wickets")
plt.xlabel('Runs')
plt.ylabel('Wickets')
plt.grid(True)
plt.show()
```

```python
#line Plot
top10 = df.sort_values(by = 'Runs',ascending=False).head(10)
plt.plot(top10['Player'],top10['Runs'],marker='^', linestyle = '--',color='red')
plt.title('top 10 players with highest scoreruns')
plt.xticks(rotation=45)
plt.xlabel('Strike Rate')
plt.ylabel('Number of Wicket')
plt.grid(True)
plt.show()
```

```python
#multiple lines within one plot
top5 = df.sort_values(by = 'Runs',ascending=False).head(5)
plt.plot(top5['Player'],top5['Runs'],marker='o',label='Runs')
top5 = df.sort_values(by = 'Wickets',ascending=False).head(5)
plt.plot(top5['Player'],top5['Wickets'],marker='s',label='Wickets')
plt.legend()
plt.title('Top 5 Players by Runs and Wickets')
plt.xlabel('Player')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
x=[1,2,3,4]
y1=[5,6,7,8]
y2=[10,11,12,13]
plt.plot(x,y1,label='Assending',color='red')
plt.plot(x,y2,label='Descending')
plt.show()
```

```python
top_teams_runs = team_stats.sort_values(by='Runs', ascending=False).head()

plt.figure(figsize=(10, 6))
plt.fill_between(top_teams_runs.index, top_teams_runs['Runs'], color='skyblue', alpha=0.4, label='Top 5 Teams Runs')

plt.title('Area Chart of Top 5 Teams by Runs')
plt.xlabel('Team')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python

top10_wickets = df.sort_values(by='Wickets', ascending=False).head(10)

plt.figure(figsize=(12, 8))
# Create a list of colors, one for each bubble
colors = plt.cm.viridis(np.linspace(0, 1, len(top10_wickets)))
plt.scatter(top10_wickets['Player'], top10_wickets['Wickets'], s=top10_wickets['Wickets']*50, alpha=0.7, c=colors)

plt.title('Top 10 Players with Most Wickets (Bubble Chart)')
plt.xlabel('Player')
plt.ylabel('Wickets')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
top5_runs_donut = df.sort_values(by='Runs', ascending=False).head(5)

# Create a pie chart
plt.pie(top5_runs_donut['Runs'], labels=top5_runs_donut['Player'], autopct='%1.1f%%', startangle=90, wedgeprops={'edgecolor': 'white'})

# Draw a circle at the center to make it a donut chart
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Top 5 Run Scorers')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
team_stats_top = team_stats.sort_values(by=['Runs', 'Wickets'], ascending=False).head(5)

fig, ax = plt.subplots()

bars1 = ax.bar(team_stats_top.index, team_stats_top['Runs'], label='Runs', color='skyblue')
bars2 = ax.bar(team_stats_top.index, team_stats_top['Wickets'], bottom=team_stats_top['Runs'], label='Wickets', color='lightcoral')

# Color the bars for the top team differently
top_team = team_stats_top.index[0]
bars1[0].set_color('darkblue')
bars2[0].set_color('darkred')


plt.title('Stacked Runs and Wickets by Top 5 Teams')
plt.xlabel('Team')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
team_wickets_top = team_stats.sort_values(by='Wickets', ascending=False).head()

plt.figure(figsize=(10, 6))

colors = ['blue', 'darkblue', 'yellow', 'red', 'orange']

# Reverse the order of teams and colors to plot from top to bottom
teams_reversed = team_wickets_top.index[::-1]
wickets_reversed = team_wickets_top['Wickets'][::-1]
colors_reversed = colors[::-1]

for i in range(len(teams_reversed)):
  plt.barh(teams_reversed[i], wickets_reversed[i], color=colors_reversed[i])

plt.title('Top 5 Teams by Wickets Taken (Gantt-like Chart)')
plt.xlabel('Total Wickets')
plt.ylabel('Team')
plt.grid(axis='x')
plt.tight_layout()
plt.show()
```

```python
top5_runs_pyramid = df.sort_values(by='Runs', ascending=False).head(5)

fig, ax = plt.subplots(figsize=(10, 6))

# We need two sets of bars for a pyramid: one positive, one negative
# We'll use runs and then mirror it for the left side of the pyramid

y_pos = np.arange(len(top5_runs_pyramid))

# Plotting the positive bars (right side of pyramid)
ax.barh(y_pos, top5_runs_pyramid['Runs'], align='center', color='skyblue')

# Mirroring the data for the negative bars (left side of pyramid)
ax.barh(y_pos, -top5_runs_pyramid['Runs'], align='center', color='lightcoral')

ax.set_yticks(y_pos)
ax.set_yticklabels(top5_runs_pyramid['Player'])
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Runs')
ax.set_title('Pyramid Chart of Top 5 Run Scorers')

# Adjust x-axis limits to make it look symmetrical and remove negative labels
max_runs = top5_runs_pyramid['Runs'].max()
ax.set_xlim([-max_runs, max_runs])
ax.set_xticks(np.linspace(-max_runs, max_runs, 5)) # Adjust number of ticks as needed
ax.set_xticklabels([str(abs(x)) for x in np.linspace(-max_runs, max_runs, 5)]) # Show absolute values

plt.grid(axis='x', linestyle='--')
plt.tight_layout()
plt.show()
```
